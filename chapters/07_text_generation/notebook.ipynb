# Chapter 07: Controlling Text Generation

LLMs generate text by predicting the next token repeatedly. **Generation parameters** let you control creativity, length, and quality.

Key parameters:
- `temperature` — randomness (0 = deterministic, 1+ = creative)
- `top_p` — nucleus sampling: only sample from top-p probability mass
- `top_k` — only sample from top-k most likely tokens
- `repetition_penalty` — penalise repeated phrases

## Setup

```python
# !pip install transformers torch
from transformers import pipeline
generator = pipeline('text-generation', model='distilgpt2')
```

## Experiment 1 — Temperature

```python
prompt = "The future of artificial intelligence is"

for temp in [0.2, 0.7, 1.2]:
    out = generator(prompt, max_new_tokens=40, temperature=temp, do_sample=True)[0]["generated_text"]
    print(f"temperature={temp}:")
    print(" ", out[len(prompt):].strip())
    print()
```

## Experiment 2 — Top-p (Nucleus Sampling)

```python
for top_p in [0.5, 0.9, 0.99]:
    out = generator(prompt, max_new_tokens=40, top_p=top_p, temperature=1.0, do_sample=True)[0]["generated_text"]
    print(f"top_p={top_p}:")
    print(" ", out[len(prompt):].strip())
    print()
```

## Experiment 3 — Repetition Penalty

```python
for penalty in [1.0, 1.3, 1.8]:
    out = generator(prompt, max_new_tokens=60, repetition_penalty=penalty, do_sample=True)[0]["generated_text"]
    print(f"repetition_penalty={penalty}:")
    print(" ", out[len(prompt):].strip())
    print()
```

## Experiment 4 — Beam Search vs Greedy vs Sampling

```python
strategies = [
    ("Greedy",        dict(do_sample=False, num_beams=1)),
    ("Beam search 4", dict(do_sample=False, num_beams=4, early_stopping=True)),
    ("Top-p 0.9",     dict(do_sample=True,  top_p=0.9, temperature=0.8)),
]
for name, kwargs in strategies:
    out = generator(prompt, max_new_tokens=40, **kwargs)[0]["generated_text"]
    print(f"[{name}]")
    print(" ", out[len(prompt):].strip())
    print()
```

## Parameter Cheat-Sheet

| Goal | Settings |
|------|----------|
| Factual / consistent | `temperature=0.2, top_p=0.9` |
| Creative writing | `temperature=0.9, top_p=0.95` |
| Long outputs, no repetition | `repetition_penalty=1.3` |
| Best single answer | `num_beams=4, do_sample=False` |
