# Chapter 7: Advanced Text Generation

> **Goal:** Master sampling strategies for text generation.


```python
from transformers import pipeline

gen = pipeline('text-generation', model='distilgpt2')
prompt = 'Scientists have discovered'
print(f'Prompt: "{prompt}"\n')

for name, kw in [('Greedy', dict(do_sample=False)), ('Temp=0.5', dict(do_sample=True,temperature=0.5)), ('Temp=1.0', dict(do_sample=True,temperature=1.0)), ('Top-p=0.9', dict(do_sample=True,top_p=0.9,temperature=0.7))]:
    r = gen(prompt, max_new_tokens=30, **kw)
    print(f'[{name}] {r[0]["generated_text"][len(prompt):].strip()}')
```
