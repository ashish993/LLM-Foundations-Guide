# Chapter 06: Prompt Engineering Patterns

Prompt engineering is the art of writing instructions that reliably guide an LLM toward the output you want — **without changing any model weights**.

We will cover 6 core patterns with live examples you can run locally.

## Setup

```python
# !pip install transformers torch
from transformers import pipeline
llm = pipeline('text-generation', model='distilgpt2', max_new_tokens=60, do_sample=False)
def ask(prompt): return llm(prompt)[0]['generated_text'][len(prompt):].strip()
```

## Pattern 1 — Zero-Shot

Ask the model directly, no examples.

```python
p = "Classify sentiment (positive/negative/neutral): 'The battery life is amazing!'
Answer:"
print(ask(p))
```

## Pattern 2 — Few-Shot

Provide 2–3 labelled examples before asking.

```python
p = """Classify sentiment:
Text: I love this product! -> positive
Text: It broke on day one. -> negative
Text: It does the job. -> neutral
Text: Absolutely fantastic experience! ->"""
print(ask(p))
```

## Pattern 3 — Chain-of-Thought (CoT)

Ask the model to *think step by step* before answering. This dramatically improves accuracy on reasoning tasks.

```python
p = """Q: Roger has 5 tennis balls. He buys 2 more cans of 3 balls each. How many balls does he have?
A: Let's think step by step.
Roger starts with 5 balls.
2 cans × 3 balls = 6 new balls.
5 + 6 = 11 balls.
Answer: 11

Q: A shop has 48 apples. They sell 15 and receive a delivery of 30 more. How many apples now?
A: Let's think step by step."""
print(ask(p))
```

## Pattern 4 — Role Prompting

Giving the model a persona focuses its style and expertise.

```python
p = "You are a senior cybersecurity analyst. Explain SQL injection in one sentence suitable for a C-level executive.
Explanation:"
print(ask(p))
```

## Pattern 5 — Output Format Control

Force structured output with explicit format instructions.

```python
p = """Extract information and respond in JSON only.
Text: "John Smith, 34, joined Acme Corp in 2019 as an engineer."
JSON: {"name": "
```

```python
print(ask(p))
```

## Pattern 6 — Self-Consistency

Sample multiple responses and pick the majority answer.

```python
from collections import Counter

prompt = "Is 1009 a prime number? Answer only yes or no.
Answer:"
responses = [ask(prompt) for _ in range(5)]
majority = Counter(r.lower()[:3] for r in responses).most_common(1)[0][0]
print("Responses:", responses)
print("Majority answer:", majority)  # 1009 IS prime
```

## Summary

| Pattern | Best for |
|---------|----------|
| Zero-Shot | Simple classification / generation |
| Few-Shot | When you have examples |
| Chain-of-Thought | Multi-step reasoning |
| Role Prompting | Tone / domain expertise |
| Format Control | APIs that need structured output |
| Self-Consistency | High-stakes decisions |
