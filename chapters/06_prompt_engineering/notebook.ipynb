# Chapter 6: Prompt Engineering

> **Goal:** Design prompts for reliable LLM behavior.


```python
# Prompt Engineering Pattern Library
patterns = {
    'Zero-Shot': 'Classify sentiment: The movie was incredible! -> POSITIVE',
    'Few-Shot': 'Examples: bad->NEGATIVE, great->POSITIVE\nNow classify: The service was slow ->',
    'Chain-of-Thought': 'Q: John has 5 apples. He gives away 2 and buys 3 more. How many? Think step by step.',
    'System Prompt': 'System: You are a Python expert. Only answer Python questions.\nUser: What is a decorator?',
    'Output Format': 'Extract name and email from: Alice Smith alice@example.com. Return JSON only.',
}

for name, example in patterns.items():
    print(f'\n[{name}]')
    print(example)

print('\nGolden rules: Be specific. Give context. Specify format. Provide examples. Iterate!')
print('\nTip: Chain-of-Thought prompting improves accuracy on math/reasoning tasks by 40-60%.')
```
