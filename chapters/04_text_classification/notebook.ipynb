## 🧠 What Is Text Classification?

Text classification assigns a label to a piece of text:

- **Sentiment analysis:** Positive / Negative / Neutral
- **Spam detection:** Spam / Not Spam
- **Topic classification:** Sports / Tech / Politics / Entertainment
- **Intent detection:** Book_flight / Cancel_order / Get_support

With pre-trained LLMs, we can do this in just a few lines — and often achieve high accuracy without any custom training!

---

## 🔄 Two Approaches

### Approach 1: Zero-shot classification (no training needed)
Ask the LLM to classify text directly via prompting.

### Approach 2: Fine-tuned classifier (BERT + classification head)
Add a classification layer on top of BERT and train it on labeled examples.

We'll demo both!

```python
from transformers import pipeline

# === APPROACH 1: Zero-shot classification ===
# No training needed — the model uses its knowledge to figure out categories

classifier = pipeline(
    'zero-shot-classification',
    model='facebook/bart-large-mnli'
)

texts = [
    'The new iPhone has an amazing camera and battery life',
    'The local football team won the championship last night',
    'Scientists discover a new treatment for Alzheimer disease',
    'The central bank raised interest rates to combat inflation',
]

category_labels = ['Technology', 'Sports', 'Health', 'Finance', 'Entertainment']

print('=== Zero-Shot Classification Results ===')
for text in texts:
    result = classifier(text, candidate_labels=category_labels)
    top_label = result['labels'][0]
    top_score = result['scores'][0]
    print(f'\nText: "{text[:60]}..."')
    print(f'  → {top_label} (confidence: {top_score:.1%})')
```

```python
from transformers import pipeline

# === Sentiment Analysis (fine-tuned model) ===
sentiment = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')

reviews = [
    'This product completely exceeded my expectations!',
    'Terrible quality, broke after one day. Never buying again.',
    'It is okay, nothing special but does the job.',
    'Absolutely brilliant! Best purchase I have made this year.',
    'The instructions were confusing and the setup took forever.',
]

print('=== Sentiment Analysis ===')
for review in reviews:
    result = sentiment(review)[0]
    label = result['label']
    score = result['score']
    emoji = '😊' if label == 'POSITIVE' else '😞'
    print(f'{emoji} [{label:8s}] ({score:.1%}) — "{review[:55]}..."')
```

```python
# === APPROACH 2: Fine-tune BERT on a custom dataset ===
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import load_dataset
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

# Load a small real-world dataset (IMDB movie reviews)
dataset = load_dataset('imdb')

print('Dataset structure:')
print(dataset)
print(f'\nSample training entry:')
print(f'  Text: {dataset["train"][0]["text"][:100]}...')
print(f'  Label: {dataset["train"][0]["label"]} (0=Negative, 1=Positive)')
```

## ✅ Chapter 4 Summary

| Method | When To Use | Accuracy | Setup Time |
|--------|------------|----------|------------|
| Zero-shot | New tasks, no labels available | Good (~70-80%) | Minutes |
| Fine-tuned | You have labeled data, need high accuracy | Excellent (~90-95%) | Hours |
| Few-shot prompting | Some examples, no training | Very good (~80-85%) | Minutes |
